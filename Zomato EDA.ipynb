```python
## Relationships
### Based on the below columns
final_df.groupby(["Aggregate rating", "Rating color", 
                  "Rating text"]).size().reset_index().rename(columns={0: "Rating Count"})
```

***Observations***
1. White - Not rated

4. Yellow - Good
5. Green - Very Good
6. Dark Green - Excellent 

```python
ratings = final_df.groupby(["Aggregate rating", "Rating color", 
                  "Rating text"]).size().reset_index().rename(columns={0: "Rating Count"})
```

```python
ratings
```

***Observations***
1. Rating from 4.5 to 4.9 --> Excellent 
2. Rating from 4.0 to 4.4 --> Very Good
3. Rating from 3.5 to 3.9 --> Good
4. Rating from 2.5 to 3.4 --> Average
5. Rating from 1.8 to 2.4 --> Poor
6. Rating 0 --> Not rated 

```python
ratings.head()
```

```python
# Plotting using barplot
import matplotlib
matplotlib.rcParams["figure.figsize"] = (12,8)
sns.barplot(x="Aggregate rating", y = "Rating Count", data=ratings)
```

```python
# Plotting using barplot using colors based on ratings.
# Mapping the colors
import matplotlib
import seaborn as sns
fig, ax = plt.subplots()
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

matplotlib.rcParams["figure.figsize"] = (15,10)
sns.barplot(x="Aggregate rating", y = "Rating Count",
            hue= "Rating color", data=ratings,
            palette=["white", "red", "orange", "yellow", "green","green"],
           ax=ax)

ax.tick_params(axis='x', colors='magenta', labelsize = 14)
ax.tick_params(axis='y', colors='magenta', labelsize = 14)

ax.set_xlabel('Aggregate Rating', color='magenta',labelpad = 14)
ax.set_ylabel('Rating Count', color='magenta',labelpad = 14)
```

***Observations***
1. Not rated count is very high
2. Maximum number of ratings are between 2.5 to 3.4


```python
## Count plot
fig, ax = plt.subplots()
fig.patch.set_facecolor('black')
ax.set_facecolor('black')

sns.countplot(x = "Rating color", data=ratings, 
              palette=["white", "red", "orange", "yellow", "green","green"],
             ax=ax)

ax.tick_params(axis='x', colors='magenta', labelsize = 14)
ax.tick_params(axis='y', colors='magenta', labelsize = 14)

ax.set_xlabel('Aggregate Rating', color='magenta',labelpad = 14)
ax.set_ylabel('Rating Count', color='magenta',labelpad = 14)
```
